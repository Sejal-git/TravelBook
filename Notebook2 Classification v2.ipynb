```python
import re
import pandas as pd
filePath= "F:/Amrita Mtech/Glisten Internship/Work/Travel Book/Sinhagad Wiki Text.txt"

#get data from txt file
file = open(filePath, "r")
doclist = [ line for line in file ]
docstr = '' . join(doclist)
sentences = re.split(r'[.!?]', docstr)
    
for i, sentence in enumerate(sentences):
        tokenized= []
        for word in sentence.split(' '):
                word = word.split('.')[0]
                word = word.lower()
                tokenized.append(word)
        sentences[i] = tokenized
#print("Number of sentences are: ", len(sentences))      
#print("\n The sentence vector is: ")
#print(sentences)

df_sentences=pd.DataFrame()
df_sentences["Sentence"]=sentences
df_sentences["id"]=df_sentences.index
i=0
for each in df_sentences["Sentence"]:
        str1=" "
        df_sentences["Sentence"][i]=str1.join(each)
        i=i+1
#print(df_sentences)

##Vector Formation : 1 Word vector
from gensim.models import word2vec
model = word2vec.Word2Vec(sentences,min_count=1)
len(model.wv.vocab)

#For averaging
import numpy as np

def buildWordVector(text):
    ##vec = np.zeros(size).reshape((1, size))
    vec=0
    count = 0
    for word in text:
        #vec =vec + model[word].reshape((1, size))
        vec =vec + model[word]
        count =count+ 1.

    if count != 0:
        vec =vec/count
    return vec

arrays = []
for sentence in sentences:
    arrays.append(buildWordVector(sentence))
#print(arrays)
#print(len(arrays))

#Cosine simillarities: 61X61 word2vec
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
import pandas as pd

A_sparse = sparse.csr_matrix(arrays)

similarities_wordvec = cosine_similarity(A_sparse)
#print('pairwise dense output:\n {}\n'.format(similarities_wordvec))

#get similarities in dataframe
df=pd.DataFrame(data=similarities_wordvec)
#print(df)

#Get Similarity sets
import pandas as pd

def get_similarityset(threshold):
    dictionary={} 
    dictionary=df.to_dict(orient='records')

    Similaritysets=[]   
    list1=[]
    SimilarityDict=[]
    for each in dictionary:
        list3=[]
        ss1=[]
        Sd1={}
        for key, value in sorted(each.items(), key=lambda item: item[1], reverse=True):
            list3.append(key)
            if (value>= threshold):
                ss1.append(value)
                Sd1.update({key:value})
        list1.append(list3)
        Similaritysets.append(ss1)
        SimilarityDict.append(Sd1)

    #print(Similaritysets)
    #print(SimilarityDict)
    #print("Lenghth of similarity sets: " ,len(SimilarityDict))
    return SimilarityDict

smilarity=[]
#print("With avg Word to vec: ")    
smilarity=get_similarityset(0.5)
print(smilarity)
```

```python
simlist1=[]
for i in range (len(smilarity)):
    for j in range (len(smilarity[i])):
        simlist2=[]
        simlist2=list(smilarity[i].values())
    simlist1.append(simlist2)
    
print(simlist1)
```

```python
#1. kmeans clustering
import pandas as pd
import numpy as np
df_kmeans=pd.DataFrame()
df_kmeans['Lists']=simlist1
df_kmeans['Numbers']=df_kmeans.index+1
avg=[]
for each in df_kmeans['Lists']:
    sum=0
    count=0
    for i in each:
        sum=sum+i
        count=count+1
    avgerage=sum/count
    avg.append(avgerage)
df_kmeans['Avg']=avg


from sklearn.cluster import KMeans
max_clusters = 3
array=np.array(df_kmeans['Avg'])
array=array.reshape(-1, 1)
kmeans = KMeans(max_clusters, random_state=0).fit(array)
print(kmeans)

#Coordinates of cluster centers with shape [n_clusters, n_features]
print(kmeans.cluster_centers_)
    
# Get the cluster labels
print(kmeans.labels_)
df_kmeans['Cluster']=kmeans.labels_
print(df_kmeans)
```

```python
#2 Time line based Classification:
#Timeline
import re
timeline=[]
for sentence in sentences:
    for l in sentence:
        match = re.match(r'.*([1-3][0-9]{3})', l)
        if match is not None:
            #print (match.group(1))
            timeline.append(match.group(1))
            
print(timeline)
```

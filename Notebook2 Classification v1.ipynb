```python
import re
import pandas as pd
filePath= "F:/Amrita Mtech/Glisten Internship/Work/Travel Book/Sinhagad Wiki Text.txt"

#get data from txt file
file = open(filePath, "r")
doclist = [ line for line in file ]
docstr = '' . join(doclist)
sentences = re.split(r'[.!?]', docstr)
    
for i, sentence in enumerate(sentences):
        tokenized= []
        for word in sentence.split(' '):
                word = word.split('.')[0]
                word = word.lower()
                tokenized.append(word)
        sentences[i] = tokenized
#print("Number of sentences are: ", len(sentences))      
#print("\n The sentence vector is: ")
#print(sentences)

df_sentences=pd.DataFrame()
df_sentences["Sentence"]=sentences
df_sentences["id"]=df_sentences.index
i=0
for each in df_sentences["Sentence"]:
        str1=" "
        df_sentences["Sentence"][i]=str1.join(each)
        i=i+1
#print(df_sentences)

##Vector Formation : 1 Word vector
from gensim.models import word2vec
model = word2vec.Word2Vec(sentences,min_count=1)
len(model.wv.vocab)

#For averaging
import numpy as np

def buildWordVector(text):
    ##vec = np.zeros(size).reshape((1, size))
    vec=0
    count = 0
    for word in text:
        #vec =vec + model[word].reshape((1, size))
        vec =vec + model[word]
        count =count+ 1.

    if count != 0:
        vec =vec/count
    return vec

arrays = []
for sentence in sentences:
    arrays.append(buildWordVector(sentence))
#print(arrays)
#print(len(arrays))

#Cosine simillarities: 61X61 word2vec
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
import pandas as pd

A_sparse = sparse.csr_matrix(arrays)

similarities_wordvec = cosine_similarity(A_sparse)
#print('pairwise dense output:\n {}\n'.format(similarities_wordvec))

#get similarities in dataframe
df=pd.DataFrame(data=similarities_wordvec)
#print(df)

#Get Similarity sets
import pandas as pd

def get_similarityset(threshold):
    dictionary={} 
    dictionary=df.to_dict(orient='records')

    Similaritysets=[]   
    list1=[]
    SimilarityDict=[]
    for each in dictionary:
        list3=[]
        ss1=[]
        Sd1={}
        for key, value in sorted(each.items(), key=lambda item: item[1], reverse=True):
            list3.append(key)
            if (value>= threshold):
                ss1.append(value)
                Sd1.update({key:value})
        list1.append(list3)
        Similaritysets.append(ss1)
        SimilarityDict.append(Sd1)

    #print(Similaritysets)
    #print(SimilarityDict)
    #print("Lenghth of similarity sets: " ,len(SimilarityDict))
    return SimilarityDict

smilarity=[]
#print("With avg Word to vec: ")    
smilarity=get_similarityset(0.5)
#print(smilarity)

simlist1=[]
for i in range (len(smilarity)):
    for j in range (len(smilarity[i])):
        simlist2=[]
        x=tuple(smilarity[i].items())
        simlist2.append(x[0])
        print(simlist2)
    simlist1.append(simlist2)
    
print(simlist1)
```

```python
'''
Using Doc2Vec
Purpose: Fetch data, create docvec vector, get cosine similarities, form similarity sets
Input: Text data
Output: Similarity sets
'''

import re
import pandas as pd
from nltk.corpus import stopwords 
filePath= "F:/Amrita Mtech/Glisten Internship/Work/Travel Book/Sinhagad Wiki Text.txt"

#get data from txt file
file = open(filePath, "r")
doclist = [ line for line in file ]
docstr = '' . join(doclist)
sentences = re.split(r'[.!?]', docstr)
 
stopwords = set(stopwords.words('english'))     
for i, sentence in enumerate(sentences):
        tokenized= []
        for word in sentence.split(' '):
            if (word not in stopwords or word==""): 
                word = word.split('.')[0]
                word = word.lower()
                tokenized.append(word)
        sentences[i] = tokenized
print("Number of sentences are: ", len(sentences))      
print("\n The sentence vector is: ")
print(sentences)

df_sentences=pd.DataFrame()
df_sentences["Sentence"]=sentences
df_sentences["id"]=df_sentences.index
i=0
for each in df_sentences["Sentence"]:
        str1=" "
        df_sentences["Sentence"][i]=str1.join(each)
        i=i+1
print(df_sentences)
```

```python
##Vector Formation :2 Doc 2 vec
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import numpy as np
import itertools

tagged_data = [TaggedDocument(d, [i]) for i, d in enumerate(sentences)]
docmodel = Doc2Vec(tagged_data, vector_size=20, window=2, min_count=1, workers=4, epochs = 100)

arrays1 = []
for i in sentences:
    vector = docmodel.infer_vector(i)
    arrays1.append(vector)
print(arrays1)

'''
arrays1 = []
for senetnce in sentences:
    list1=[]
    list1=docmodel[senetnce]
    print(senetnce)
    print(docmodel[senetnce])
    flattened_list = [y for x in list1 for y in x]
    print(flattened_list)
    arrays1.append(flattened_list)

for l in range(0,61):
    list1=[]
    list1=list(itertools.chain.from_iterable( arrays1[l]))
    arrays1[l]=list1

print(arrays1)
print(len(arrays1))   
'''
```

```python
#Cosine simillarities: 61X61 Doc2Vec
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
import pandas as pd

A_sparse1 = sparse.csr_matrix(arrays1)
similarities_docvec = cosine_similarity(A_sparse1)
print('pairwise dense output:\n {}\n'.format(similarities_docvec))

#get similarities in dataframe
df1=pd.DataFrame(data=similarities_docvec)
print(df1)
```
